```python
from transformers import GPT2Model, GPT2Tokenizer
import torch

# Load GPT-2 pre-trained and its tokenizer
model_name = "gpt2"
model = GPT2Model.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
# Example text
text = "This is a sample sequence to determine the output dimension of the GPT final hidden state."

# Tokenize sequence
inputs = tokenizer(text, return_tensors="pt").to(device)

# Extract hidden state of the last layer
with torch.no_grad():
    outputs = model(**inputs)

# Print shape of di h_n
h_n = outputs.last_hidden_state
print(f"Shape of h_n: {h_n.shape}")   # (batch_size]

# Get transposed Embedding matrix
W_e_T = model.wte.weight.T

# Print shape of transposed Embedding matrix
print(f"Shape of W_e^T: {W_e_T.shape}")

# Calculate product h_n * W_e^T
logits = torch.matmul(h_n, W_e_T)

# Print logit shape:
print(f"Shape of h_n * (W_e)^T: {logits.shape}")
```
